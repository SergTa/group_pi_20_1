## В целом понятно, погнали обучать

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
from sklearn.model_selection import train_test_split
```

```python
def build_result_and_print_metrics(model_name, y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precission = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    print("accuracy", accuracy)
    print("precision", precission)
    print("recall", recall)
    print("f1", f1)
    return {
        'Model_Name': model_name,
        'accuracy': accuracy,
        'precision': precission,
        'recall': recall,
        'f1': f1,
    }
```

```python
model_results = []
```

```python
dfx = df_all[features_with_target]
dfx.head()
```

```python
X = dfx.drop(columns=[target])
y = dfx[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model_lr = LogisticRegression(max_iter=2000, random_state=42)
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)

rslt = build_result_and_print_metrics('LogisticRegression', y_test, y_pred)
model_results.append(rslt)
```

```python
model_dtc = DecisionTreeClassifier(max_depth=20, random_state=42)
model_dtc.fit(X_train, y_train)
y_pred = model_dtc.predict(X_test)

rslt = build_result_and_print_metrics('DecisionTreeClassifier', y_test, y_pred)
model_results.append(rslt)
```

```python
results_df = pd.DataFrame(model_results)
results_df
```

```python
results_df.plot(
    x="Model_Name",
    y=["accuracy", "precision", "recall", "f1"],
    kind="bar",
    figsize=(10, 6),
)
plt.xticks(rotation=0, ha="left", fontsize=12)
plt.legend(title="Metrics", fontsize=12)
plt.tight_layout()
```
